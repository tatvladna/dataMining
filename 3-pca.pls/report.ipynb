---

## Цель лабораторной работы

` Уменьшение размерности дескрипторного пространства с использованием метода главных компонент (PCA) для анализа и визуализации набора данных, а также оценки вклада отдельных дескрипторов в объясненную дисперсию.`


----------------------

## Задачи лабораторной работы

1. Разделить представленный набор данных на обучающую и тестовую выборки в соотношении 80:20.

2. Создать и обучить конвейер, включающий генерацию дескрипторов с помощью библиотеки RDKit, нормализацию и применение метода главных компонент (PCA) для сохранения 95% объясненной дисперсии.

3. Сохранить обученный конвейер и использовать его для генерации дескрипторов тестовой выборки.

4. Визуализировать обучающую и тестовую выборки в пространстве главных компонент.

5. Оценить вклад отдельных дескрипторов с помощью графика нагрузок.

6. Подобрать оптимальное количество главных компонент для сохранения более 95% объясненной дисперсии.

-------------

Поэтапное изложение процедуры

На первом этапе работы проводится разделение представленного набора данных на обучающую и тестовую выборки в соотношении 80:20. Это разделение необходимо выполнять заранее, чтобы избежать утечки данных и корректно оценить производительность модели на независимом наборе данных.

Далее создается и обучается конвейер (sklearn.pipeline.Pipeline), который включает несколько этапов обработки данных. В первую очередь производится генерация дескрипторов с использованием библиотеки RDKit. В качестве характеристик молекул используются такие параметры, как число тяжелых атомов, число NHOH групп, число NO групп, число акцепторов и доноров водородной связи, число гетероатомов, число способных к вращению связей, число валентных электронов, число ароматических циклов, число алифатических гетероциклов, общее число циклов в молекуле, молекулярная масса, коэффициент распределения в системе "вода – 1-октанол", площадь топологической полярной поверхности и молярная рефракция. После генерации дескрипторов выполняется их нормализация, что позволяет привести признаки к единому масштабу и улучшить работу модели. Далее к нормализованным данным применяется метод главных компонент (PCA), который позволяет сократить размерность пространства признаков, сохраняя при этом 95% объясненной дисперсии.

После обучения конвейер сохраняется в pickle-файле для последующего использования. С его помощью проводится генерация дескрипторов для тестовой выборки. Полученные данные преобразуются в pandas DataFrame и сохраняются в pickle и csv файлах для дальнейшего анализа.

Визуализация набора данных и оценка вкладов дескрипторов

Для анализа структуры данных проводится визуализация обучающей выборки с использованием метода главных компонент. Строятся двумерные и трехмерные графики, на которых отображаются объекты в пространстве главных компонент. Определяется процент объясненной дисперсии и проводится раскраска точек в зависимости от класса активности. Это позволяет оценить степень разделимости классов в новом пространстве признаков.

Далее выполняется проекция тестовой выборки на ранее построенные графики. Это позволяет оценить, насколько распределение тестовых данных соответствует распределению обучающих данных и проверить, сохраняется ли выявленная структура в новом наборе данных.

Для более детального анализа строится график нагрузок на обучающей выборке. С его помощью оцениваются вклады отдельных дескрипторов в формирование главных компонент, а также их корреляция друг с другом. Это позволяет определить, какие признаки оказывают наибольшее влияние на распределение данных и какие из них могут быть избыточными.

На заключительном этапе работы проводится подбор оптимального количества главных компонент, позволяющего сохранить более 95% объясненной дисперсии. Это позволяет минимизировать размерность признакового пространства, сохраняя при этом всю важную информацию о данных.

## Результаты работы (+ таблицы, графики, иллюстрации?, если требуются)

В ходе работы было проведено уменьшение размерности дескрипторного пространства с использованием метода главных компонент, что позволило сохранить 95% объясненной дисперсии при уменьшении количества признаков.

На тренировочных данных использование MinMaxScaler показало, что график нагрузок между первой и второй главной компонентой выявил наибольший вклад числа алифатических гетероциклов. Четыре компоненты объясняли 95% дисперсии, а добавление пятой позволяло достичь 100%. В случае применения StandardScaler потребовалось уже пять главных компонент для сохранения 95% дисперсии, а шесть компонент полностью объясняли дисперсию данных. Анализ нагрузок показал, что наибольшее влияние на первую и вторую главные компоненты оказало число NHOH групп.

На тестовой выборке результаты были схожими. Для MinMaxScaler четыре компоненты объясняли 95% дисперсии, а пять – 100%. Число алифатических гетероциклов продолжало вносить основной вклад в образование первой и второй компонент. Для StandardScaler также четыре компоненты обеспечивали 95% объясненной дисперсии, а пять – 100%. График нагрузок подтвердил, что число NHOH групп также оставалось главным фактором формирования первых двух главных компонент.
